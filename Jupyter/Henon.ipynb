We use the powerful package [DynamicalSystems.jl](https://juliadynamics.github.io/DynamicalSystems.jl) package for genereating 1 million points with the HÃ©non map and show the graphical result.  

```julia
using DynamicalSystems
using DynamicalSystemsBase.Systems
using StaticArrays # only necessary when defining a system


he = DynamicalSystemsBase.Systems.henon([0.,0.0]; a = 1.4, b = 0.3)
```

```julia
traj = trajectory(he, 100000);
```

```julia
using GR
shade(traj[:,1],traj[:,2], colormap=-2, xlabel="x", ylabel="y")
```
